```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##SENTENCES
sent= {"Hello, my dog is cute", 
       "Hello, my cat is cute",
       "Hello, my dog is very cute",
       "your videos are very good",
       "i am very good boy",
       "i am good developer",
       "the glass of milk"}    

sent
```

```python
### define vocab size
voc_size=10000
```

```python
###one hot representation
one_hot_repr = [one_hot(words, voc_size) for words in sent]
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential

```

```python
import numpy as np
```

```python
from tensorflow.keras.utils import pad_sequences

sent_length = 15

embedded_docs = pad_sequences(
    one_hot_repr,
    padding='post',
    maxlen=sent_length
)

print(embedded_docs)
```

```python
dim=10
```

```python
#   feature representation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

model = Sequential()

model.add(
    Embedding(
        input_dim=voc_size,
        output_dim=10,
        input_length=sent_length
    )
)

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

# Recreate the model only if it is not already defined in this session
if 'model' not in globals():

    model = Sequential()
    model.add(Embedding(
        input_dim=voc_size,
        output_dim=10,
        input_length=sent_length
    ))
    model.compile(optimizer='adam', loss='mse')

# Predict on the first sentence as a batch of one
model.predict(embedded_docs[0:1])
```
